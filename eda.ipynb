# Univariate Analysis

```python
numerical_cols = ['Age', 'Annual Income', 'Health Score', 'Previous Claims',
                  'Vehicle Age', 'Credit Score', 'Insurance Duration',
                  'Number of Dependents', 'Premium Amount']

fig, axes = plt.subplots(3, 3, figsize=(14, 16))
axes = axes.flatten()

for i, col in enumerate(numerical_cols):
    sns.histplot(train_df[col], kde=True, color = 'red', ax=axes[i])
    axes[i].set_title(f'{col} Distribution')

plt.tight_layout()
plt.show()
```

```python
categorical_cols = ['Gender', 'Marital Status', 'Education Level', 'Occupation',
                    'Location', 'Policy Type', 'Customer Feedback',
                    'Smoking Status', 'Exercise Frequency', 'Property Type']

fig, axes = plt.subplots(5, 2, figsize=(14, 20))
axes = axes.flatten()

for i, col in enumerate(categorical_cols):
    order = train_df[col].value_counts().index
    sns.countplot(x=col, data=train_df, order=order, ax=axes[i])
    axes[i].set_title(f'{col} Distribution')
    axes[i].tick_params(axis='x', rotation=30)

plt.tight_layout()
plt.show()
```

# Outlier Detection (Boxplots for numerical)

```python
fig, axes = plt.subplots(3, 3, figsize=(14, 16))
axes = axes.flatten()

for i, col in enumerate(numerical_cols):
    sns.boxplot(y=train_df[col], ax=axes[i])
    axes[i].set_title(f'{col} — Boxplot')

plt.tight_layout()
plt.show()
```

```python
train_df[numerical_cols].skew().sort_values(ascending=False)
```

# Bivariate Analysis (Feature vs Target)

```python
fig, axes = plt.subplots(3, 3, figsize=(14, 16))
axes = axes.flatten()

for i, col in enumerate(numerical_cols):
    sns.scatterplot(x=train_df[col], y=train_df['Premium Amount'], alpha=0.3, ax=axes[i])
    axes[i].set_title(f'{col} vs Premium Amount')

plt.tight_layout()
plt.show()
```

```python
corr = train_df[numerical_cols].corr()

plt.figure(figsize=(12, 8))
sns.heatmap(corr, annot=True, fmt='.2f', cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
for col in categorical_cols:
    plt.figure(figsize=(8, 4))
    sns.boxplot(x=col, y='Premium Amount', data=train_df)
    plt.title(f'{col} vs Premium Amount')
    plt.xticks(rotation=30)
    plt.show()
```

# Date Column Analysis (Policy Start Date)

```python
train_df['Policy Start Date'] = pd.to_datetime(train_df['Policy Start Date'])
train_df['Policy Year'] = train_df['Policy Start Date'].dt.year
train_df['Policy Month'] = train_df['Policy Start Date'].dt.month
train_df['Policy Day'] = train_df['Policy Start Date'].dt.day

# Distribution by year
sns.countplot(x='Policy Year', data=train_df)
plt.title('Policies by Year')
plt.show()

# Monthly trend in premium
train_df.groupby('Policy Month')['Premium Amount'].mean().plot(kind='bar')
plt.title('Avg Premium by Month')
plt.show()

# Daily trend in premium
train_df.groupby('Policy Day')['Premium Amount'].mean().plot(kind='bar')
plt.title('Avg Premium by Day')
plt.show()
```
